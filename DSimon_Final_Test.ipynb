Section 2: Data Cleaning
- Identify missing values using isnull().sum()
- Handle missing values:
* Numerical → mean/median
* Categorical → mode
- Remove duplicates
- Re-validate dataset cleanliness

```python
#IDENTIFY MISSING VALUES USING ISNULL.SUM
#missingvalues Shows - 8 missing values in the following columns
# in command at incident, complainant ethnicity, complainant gender, complainant age incident, allegation, precinct,
#contact reason, outcome description,
print('Missing values per column:')
print(df.isna().sum())
```

```python
#2B HANDLE MISSING VALUES
###2B1. Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)
df.columns
```

```python
missing_by_column = df.isna().sum().sort_values(ascending=False).to_frame(name='Missing Values')
missing_by_column
```

```python
clean_df = df.copy()

numeric_cols = clean_df.select_dtypes(include=[np.number]).columns
categorical_cols = clean_df.select_dtypes(include=['object']).columns

# Impute numeric columns with median
for col in numeric_cols:
    clean_df[col] = clean_df[col].fillna(clean_df[col].median())

# Impute categorical columns with mode
for col in categorical_cols:
    mode_series = clean_df[col].mode(dropna=True)
    if not mode_series.empty:
        clean_df[col] = clean_df[col].fillna(mode_series[0])

print('Numeric columns imputed with median:')
print(list(numeric_cols))
print('\nCategorical/object columns imputed with mode:')
print(list(categorical_cols))
```

```python
#Verify if the new data is fully clean
remaining_missing = clean_df.isna().sum().sum()
print(f'Remaining missing values after imputation: {remaining_missing}')

clean_df.isna().sum().sort_values(ascending=False)
```

```python
#Descriptive Statistics – Categorical Column
def show_category_distribution(col_name):
    counts = df[col_name].value_counts()
    percents = df[col_name].value_counts(normalize=True) * 100
    display(pd.DataFrame({'count': counts, 'percent': percents.round(2)}))

for col in ['mos_ethnicity', 'complainant_ethnicity', 'contact_reason', 'fado_type', 'allegation', 'board_disposition']:
    print(f'--- {col} ---')
    show_category_distribution(col)
    print()
```
